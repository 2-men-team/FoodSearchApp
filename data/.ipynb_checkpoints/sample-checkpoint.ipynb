```python
import pandas as pd
import numpy as np
```

```python
menu = pd.read_csv("Menu.csv")
```

```python
dishes = pd.read_csv("Dish.csv")
```

```python
menu = menu.drop_duplicates(subset=['sponsor'], keep='first')
menu = menu.dropna(subset=['sponsor'])
dishes = dishes.dropna(subset=['lowest_price', 'name'])
```

```python
menu.info()
```

```python
class Restaurant:
    def __init__(self, name, location, lat, long):
        self.name = name
        self.location = location
        self.lat = lat
        self.long = long
        self.dishes = []
        
    def add(self, dish):
        self.dishes.append(dish)
        
class Dish:
    def __init__(self, name, price):
        self.name = name
        self.price = price
```

```python
rests = [Restaurant(row['sponsor'], row['location'], 0, 0) for _, row in menu.iterrows()]
```

```python
coords = pd.read_csv("coords.csv")

for x in range(len(rests)):
    print(x, len(rests))
    rests[x].lat = coords.iloc[[x]]['Latitude'][x]
    print(rests[x].lat)
    rests[x].long = coords.iloc[[x]]['Longitude'][x]
    print(rests[x].long)
```

```python
dishes = [Dish(row['name'], (row['lowest_price'] + row['highest_price'])/2.) for _, row in dishes.iterrows()]
```

```python
import random
np.random.seed(42)
random.seed(42)

length = len(rests)

for dish in dishes:
    rests[random.randint(0, length - 1)].add(dish)
```

```python
def prep(string):
    chars = ['"', "'", ',', '.', ';', '!', '?', '&', '$', '@', '|', ':']
    for char in chars:
        string = string.replace(char, "")
    return string.lower()

def to_csv(data, path='data_delim.csv'):
    with open(path, 'w') as f:
        for i in range(len(data) - 1):
            print(i)
            rest = data[i]
            for dish in rest.dishes:
                f.write(prep(str(rest.name)) + ':' + prep(str(rest.location)) + ':' + str(rest.lat) + ':' + str(rest.long) + ':' + prep(str(dish.name)) + ':' + str(dish.price) + ':' + '\n')
        f.write(prep(str(rest.name)) + ':' + prep(str(rest.location)) + ':' + prep(str(dish.name)) + ':' + str(dish.price))
```

```python
to_csv(rests)
```
